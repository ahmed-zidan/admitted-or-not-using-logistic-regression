```python
#set inputs and outputs
data.insert(0,"ones",1)

cols = data.shape[1]
x = data.iloc[: , 0:cols-1]
y = data.iloc[:,cols-1:cols]
```

```python
print(x.head())
print(y.head())
```

```python
x = np.array(x.values)
y = np.array(y.values)
theta = np.zeros(cols-1)
```

```python
print(x)
print(y)
print(theta)
```

```python
def cost(theta , x, y):
    theta = np.matrix(theta)
    x = np.matrix(x)
    y = np.matrix(y)
    z = sigmoid(x*theta.T)
    first = np.multiply(-y , np.log(z))
    second = np.multiply(1-y,np.log(1 - z))
    
    return np.sum(first - second)/len(x)
    
    
```

```python
currCost = cost(theta,x,y)
print(currCost)
```

```python
def gradiant(theta , x, y):
    
    theta = np.matrix(theta)
    x = np.matrix(x)
    y = np.matrix(y)
    
    par = int(theta.ravel().shape[1])
    grad = np.zeros(par)
    
    error = sigmoid(x * theta.T) - y
    
    for i in range(par):
        temp = np.multiply(error , x[:,i])
        grad[i] = np.sum(temp) / len(x)
        
    return grad
    
    
```

```python
import scipy.optimize as opt

result = opt.fmin_tnc(func = cost , x0 = theta , fprime = gradiant , args = (x,y))
```

```python
print(result)
```

```python
minCost = cost(result[0] , x,y)
print(minCost)
```

```python
def predict(theta , x):
    prob = sigmoid(x * theta.T)
    return [1 if i >= 0.5 else 0 for i in prob]

theta_min = np.matrix(result[0])
predictions = predict(theta_min,x)

print(predictions)
```
